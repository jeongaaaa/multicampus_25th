#1. fillna( ): 결측값을 채워주는 함수

```python
df.info()
```

키에 2개의 데이터가 누락, 그룹에 1개의 데이터가 누락된 것을 확인 가능

먼저, 키가 없는 사람은 존재할 수 없으므로 데이터가 잘못되어 있다는 것을 표기 하기 위하여, **누락된 데이터를 -1로 채우기**

```python
df['키']
```

```python
# NaN값을 찾아서 NaN인 경우 그 데이터를 특정값(-1)으로 채워주는 함수
df['키'].fillna(-1)
```

NaN값이 -1로 바뀌었음

이때, **키의 NaN 값을 채워준 다음 유지시키려면 inplace=True 옵션을 주거나 fillna로 채워준 값을 다시 대입**해줘야함

```python
df2 = df.copy()
```

```python
df2
```

```python
df2['키'].fillna(-1, inplace=True)
```

```python
df2['키']
```

이제 유지가 됨. but, 강사님은 fillna로 채워준 값을 다시 대입 해 주는 방식(아래의 방식)을 추천함! 

```python
df2 = df.copy()
```

```python
df2['키']
```

```python
df2['키'] = df2['키'].fillna(-1)
```

```python
df2['키']
```

#2. 빈값(NaN)이 있는 행을 제거

```python
pd.__version__
```

```python
df = pd.read_csv('https://bit.ly/ds-korean-idol')
```

```python
df.info()
```

```python
df
```

```python
df.dropna()
```

dropna()는 몇 가지 옵션 추가 가능

##2-1. axis (열/행을 drop)

axis=0은 행을 drop

```python
df = df.dropna(axis=0)
```

axis=1은 열을 drop

```python
df = df.dropna(axis=1)
```

```python
df
```

```python
# dropna()와 같은 함수 실행시 인덱스가 변경될때, 다시 인덱스값을 부여할 때 
df.reset_index(drop=True)
```

##2-2. how옵션
- 'any': 한개라도 있는 경우 드랍
- 'all'은 모두 NaN인 경우 드랍

```python
df = pd.read_csv('https://bit.ly/ds-korean-idol')
```

```python
df
```

```python
# default : how='any'
df.dropna()
```

```python
df.dropna(axis=0, how='any')
```

```python
df.dropna(axis=0, how='all')
```

```python
import numpy as np
```

```python
# 10번째행에 전체가 NaN인 행이 추가
df.iloc[10] = np.nan
```

```python
df
```

```python
df.dropna(axis=0, how='all')
```

#3. 중복된 값 제거 (drop_dulicates)

```python
df = pd.read_csv('https://bit.ly/ds-korean-idol')
```

```python
df
```

##3-1. column의 중복값 제거

```python
df['키']
```

```python
df['키']=df['키'].fillna(178.0)
```

```python
# 첫번째 데이터만 남기고 나머지는 다 삭제
df['키'].drop_duplicates()
```

keep 옵션으로 유지하고 싶은 데이터를 선택할 수 있습니다. keep: 'first' / 'last'

```python
df = pd.read_csv('https://bit.ly/ds-korean-idol')
```

```python
df['키']
```

```python
df['키'].drop_duplicates(keep='last')
```

```python
df['키'] = df['키'].drop_duplicates(keep='last')
```

```python
df
```

##3-2. 행 전체 제거

```python
df.drop_duplicates('그룹')
```

```python
df.drop_duplicates('그룹', keep='last')
```

#4. Drop - column / row 제거

```python
df = pd.read_csv('https://bit.ly/ds-korean-idol')
```

```python
df.head()
```

##4-1. column 제거
: drop()을 활용하여 column을 제거 가능. column을 제거할 때는 axis=1 옵션

```python
df.drop('그룹', axis=1)
```

복수의 column을 제거하고 싶을 때는 list로 지정

```python
df.drop(['그룹', '소속사'], axis=1)
```

##4-2. row 제거
: drop()을 활용하여 row 제거 가능. row를 제거하고 싶을 때는 제거하고 싶은 index와 axis=0 옵션

```python
df.drop(3, axis=0)
```

복수의 row를 제거하고 싶을때는 list 지정

```python
df.drop([3, 5], axis=0)
```
