#2. DataFrame 병합하기 (merge)

concat과 merge는 단순히 합치기만 하는 것인지 아니면 특정 기준(index)으로 합치는 것인지에 따라 나뉨

- concat: row나 column 기준으로 단순하게 이어 붙이기
- merge: 특정 고유한 키(unique id) 값을 기준으로 병합하기

```python
df = pd.read_csv('https://bit.ly/ds-korean-idol')
df2 = pd.read_csv('https://bit.ly/ds-korean-idol-2')
```

```python
df.head()
```

```python
df2.head()
```

df와 df2는 '이름' 이라는 column이 겹침

→ '이름'을 기준으로 두 DataFrame을 병합 가능


**pd.merge(left, right, on='기준column', how='left')**

- left와 right에는 병합할 두 DataFrame을 대입
- on에는 병합의 기준이 되는 column을 대입
- how 에는 'left', 'right', 'inner', 'outer' 라는 4가지의 병합 방식중 한가지를 택해서 적어줌

##2-1. left, right 방식

먼저, 다음과 같이 df2 에서 5개의 데이터를 고의적으로 drop 해줌

```python
df2
```

```python
df_right = df2.drop([1, 3, 5, 7, 9], axis=0)
```

```python
df_right = df_right.reset_index(drop=True)
```

```python
df_right
```

```python
df
```

```python
# 이름이나 데이터를 보고 데이터프레임을 합쳐주는것이 아니라
# 기계적으로 같은 행이나 열을 맞춰서 merge해주는 함수
pd.concat([df, df_right], axis=1)
```

```python
df
```

병합할 2개의 DataFrame의 행의 갯수가 맞지 않음

→ 'left' 옵션을 부여하면, left DataFrame에 키 값이 존재하면 해당 데이터를 유지하고, 병합한 right DataFrame의 값으 NaN이 대입 됩니다.

```python
# 서로 정렬 순서가 다른 2개의 데이터프레임을 합칠때 merge()함수를 사용
# 정렬할때 참조하는 '이름'으로 이용해서 2 데이터 프레임을 합친다.
# 정렬기준 순서는 왼쪽 기준으로 한다
# merge를 사용하면 오른쪽의 '이름' column은 삭제
pd.merge(df, df_right, on='이름', how='left')
```

반대로 'right' 옵션을 부여하면 right DataFrame을 기준으로 병합하게 됨

현재, left DataFrame이 더 많은 데이터를 보유하고 있기 때문에 right를 기준으로 병합하면 DataFrame 사이즈가 줄어들 것

```python
pd.merge(df, df_right, on='이름', how='right')
```

위의 예제에서, right 를 기준으로 병합한 결과 DataFrame 사이즈가 줄어들었음

##2-2. inner, outer 방식

- inner 방식은 두 DataFrame에 모두 키 값이 존재하는 경우만 병합
- outer 방식은 하나의 DataFrame에 키 값이 존재하는 경우 모두 병합
- outer 방식에서는 없는 값은 NaN으로 대입됨

```python
df
```

```python
df_right
```

```python
# 2개의 데이터 프레임을 합쳤을때 두 데이터 프레임중에 결측치가 있는 행은 삭제
# 오른쪽 데이터 프레임의 '이름' column의 결측치가 있는 경우
pd.merge(df, df_right, on='이름') #, how='inner')
```

```python
pd.__version__
```

```python
# 2개의 데이터프레임중에 하나라도 데이터가 존재하면 그 행은 합쳤을때 존재함
pd.merge(df, df_right, on='이름', how='outer')
```

##2-3. column명은 다르지만 동일한 성질의 데이터인 경우?

```python
df
```

```python
df_right
```

```python
df_right.columns = ['성함', '연봉', '가족수']
```

```python
df_right
```

```python
df
```

df와 df_right를 병합하려고 했더니, df에서는 '이름', df_right에서는 '성함'으로 표기되어 기준이 되는 column을 지정할 수 없음

→ 이런 경우 **left_on, right_on** 옵션 사용

```python
pd.merge(df, df_right, left_on='이름', right_on='성함', how='outer')
```
