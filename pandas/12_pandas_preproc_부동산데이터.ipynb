#3. 분양가격 column을 int 타입으로 변환 (실전 전처리)

```python
df['분양가격'].astype(int)
```

데이터 에러 없어지도록 전처리하기

###strip() 활용하여 데이터의 공백 없애기

[tip] column의 문자열에 strip을 실행하고자 할 때는 **str.strip()**를 실행해줘야 합니다

[예시] df['분양가격'].str.strip()

```python
# 분양가격 컬럼에서 공백 '  '만 출력
df.loc[df['분양가격'] == '  ']
```

```python
df['분양가격'] = df['분양가격'].str.strip()
```

```python
df.loc[df['분양가격'] == '']
```

###빈 공백이 있는 데이터에 0 넣어주기

**loc 활용**

```python
df.loc[ df['분양가격'] == '', '분양가격'] = 0
```

```python
df['분양가격'].astype(int)
```

아직도 에러 발생 → NaN값 처리해줘야함

###NaN값은 fillna로 채워주기

```python
df['분양가격'] = df['분양가격'].fillna(0)
```

```python
df['분양가격'].astype(int)
```

```python
df.loc[df['분양가격'] == '6,657']
```

위의 행과 같이 ","가 들어있는 데이터에 의한 에러 발생

###콤마 제거해주기

```python
# replace함수 사용

df['분양가격'] = df['분양가격'].str.replace(',', '')
```

```python
df.iloc[2125]

#변환된 것 확인
```

```python
df['분양가격'].astype(int)
```

NaN값이 다시 생겨서 오류 발생 → fiina로 다시 처리해주기

```python
df['분양가격'] = df['분양가격'].fillna(0)
```

```python
df['분양가격'].astype(int)
```

###NaN값 0으로 채워주기

#6. 지역명 별로 평균 분양가격 확인

```python
df.groupby('지역명')['분양가격'].mean()
```

###분양가격이 100보다 작은 행 제거해보기

```python
df.loc[ df['분양가격'] < 100]
```

[tip] 특정 조건에 만족하는 행을 제거하고자 할 때는

1. **index를 list로** 가져온다
2. **drop을 활용**하여 행을 제거한다

```python
# index 가져오기
idx = df.loc[ df['분양가격'] < 100].index
```

```python
idx
```

```python
df
```

```python
# index를 행 기준으로 drop

df = df.drop(idx, axis=0)
```

```python
df.count()
```

df의 행의 갯수가 줄어든것이 보이시나요?

df.count() 값이 4185개가 출력이 되어야 합니다!

다시 한 번 지역명으로 group을 묶어 분양가격을 확인해 보겠습니다.

```python
df.groupby('지역명')['분양가격'].mean()
```

#8. 피벗 테이블 활용

- 행인덱스: 연도
- 열인덱스: 규모구분
- 값: 분양가격

```python
pd.pivot_table(df, index='연도', columns='규모구분', values='분양가격')
```

#9. 연도별, 규모별 가격 알아보기 (multi-index)

```python
df.groupby(['연도', '규모구분'])['분양가격'].mean()
```

예쁘게 출력이 안되어서 보기가 힘들때는 pd.DataFrame()으로 한 번 더 감싸주면 됨

```python
pd.DataFrame(df.groupby(['연도', '규모구분'])['분양가격'].mean())
```
