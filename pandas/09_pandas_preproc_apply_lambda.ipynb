##1-1. (목표) 남자/ 여자의 문자열 데이터로 구성된 '성별' column을 1 / 0 으로 바꿔보기

```python
df = pd.read_csv('http://bit.ly/ds-korean-idol')
```

```python
df
```

남자: 1 여자: 0 기타: -1

**[주의] 반드시 return 값이 존재해야 함**

```python
# 먼저 apply에 활용할 함수를 정의

def male_or_female(x):
    if x == '남자':
        return 1
    elif x == '여자':
        return 0
```

```python
#.map()이라는 함수와 유사
#둘의 차이는 .map()은 앞에 입력되는 데이터타입이 list 등의 python data type으로 다양하지만, 
#apply 함수는 pandas의 series 등의 데이터타입을 다루기에 최적화 되어있음

df['성별_NEW'] = df['성별'].apply(male_or_female)
```

```python
#과제
#위의 male_or_female 함수를 lambda함수로 바꾸기
# https://wikidocs.net/24

# 람다함수의 예(1)
add = lambda a, b: a+b

result = add(3, 4)
print(result)
```

```python
#람다함수의 예(2)
score = lambda x: 'A등급' if x >= 80 else 'B등급'
```

```python
male_or_female = lambda x: 1 if x == '남자' else 0
```

##1-2. (목표) cm당 브랜드 평판지수 구해보기 (브랜드평판지수/ 키)

키: 178 브랜드평판지수: 99000

값: 99000 / 178

```python
# 먼저 apply에 활용할 함수를 정의하기

def cm_to_brand(df):
    value = df['브랜드평판지수'] / df['키']
    return value
```

```python
# apply를 통해 cm_to_brand함수의 입력으로 df 데이터프레임 전체 전달
# df를 전달할때는 axis=1이어야 함

df.apply(cm_to_brand, axis=1)
```

#2. lambda 함수의 적용
- lambda는 1줄로 작성하는 간단 함수식
- return을 별도로 명기하지 않음

```python
f = lambda x: 1 if x == '남자' else 0
```

```python
df['성별'].apply(f)
```

실제로는 **간단한 계산식을 적용하려는 경우**에 많이 사용

```python
df['키/2'] = df['키'].apply(lambda x: x / 2)
```

```python
df.head()
```

```python
df['키'].apply(lambda x: x ** 2)
```

apply에 함수식을 만들어서 적용해주는 것과 동일하기 때문에, **복잡한 조건식은 함수로 간단한 계산식은 lambda로 적용**할 수 있음
